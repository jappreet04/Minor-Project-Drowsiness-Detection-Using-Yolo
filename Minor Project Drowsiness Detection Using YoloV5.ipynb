```python
cap = cv2.VideoCapture('biking_mountainbiking_tour_road_bike_947.mp4')
while cap.isOpened():
    ret, frame = cap.read()
    
    
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import uuid
import os
import time
```

```python
IMAGES_PATH = os.path.join( 'Untitled Folder 1', 'images') 
labels = ['awake', 'drowsy']
number_imgs = 20
```

```python
cap = cv2.VideoCapture(0)
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
```

```python
cap = cv2.VideoCapture(0)

for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
   
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        
        
        ret, frame = cap.read()
        
       
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg')
        
       
        cv2.imwrite(imgname, frame)
        
       
        cv2.imshow('Image Collection', frame)
        
       
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
